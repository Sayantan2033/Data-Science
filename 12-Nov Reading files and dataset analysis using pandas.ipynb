### Write a pandas program and define a function that returns every record with title starting with "Once", "once" and "ONCE":

```python
movies.title[78] = "once Upon a Time in America"
movies.title[402] = "ONCE Upon a Time in America"
```

```python
movies[movies.title.str.startswith("Once")]
```

```python
def once(s):
    if s[0:4]=="Once" or s[0:4]=="ONCE" or s[0:4]=="once":
        return True
    else:
        return False

m = movies.title.apply(once) #Sends the title of each record one-by-one
movies[m]
```

#### Using regular expression:

```python
import re

def check(s):
    if re.search("^once",s,re.I):
        return True
    else:
        return False
    
movies[movies.title.apply(check)]
```
