# Line plots

## Simple line plot:

```python
import matplotlib.pyplot as plt

x = [1,2,3]

y_1 = [50,60,70]
y_2 = [20,30,40]

plt.plot(x, y_1, marker="x")
plt.plot(x, y_2, marker="^")

plt.xlim([0, len(x)+1])
plt.ylim([0, max(y_1+y_2)+10])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Simple line plot")
plt.legend(["Sample 1","Sample 2"], loc="upper left")

plt.grid() #Making a grid
plt.show()
```

## Line plots with error bars:

```python
import matplotlib.pyplot as plt

x = [1,2,3]

y_1 = [50,60,70]
y_2 = [20,30,40]

#Errors
y_1_err = [4.3,4.5,2.0]
y_2_err = [2.3,6.9,2.1]

x_labels = ["x1","x2","x3"]

#Adding error bars
plt.errorbar(x, y_1, yerr=y_1_err, fmt="-x")
plt.errorbar(x, y_2, yerr=y_2_err, fmt="-^")

#Adding x-ticks
plt.xticks(x, x_labels)

plt.xlim([0, len(x)+1])
plt.ylim([0, max(y_1+y_2)+10])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line plot with error bars")
plt.legend(["Sample 1","Sample 2"], loc="upper left")

plt.grid() #Making a grid
plt.show()
```

## Line plot with X-axis labels and log-scale:

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3]

y_1 = [0.5,7.0,60.0]
y_2 = [0.3,6.0,30.0]

x_labels = ["x1","x2","x3"]

plt.plot(x, y_1, marker="x")
plt.plot(x, y_2, marker="^")

plt.xlim([0, len(x)+1])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

#Making it log scale
plt.yscale("log")

plt.title("Line plot with x-axis labels and log-scale")
plt.legend(["Sample 1","Sample 2"], loc="upper left")

plt.grid() #Making a grid
plt.show()
```

## Gaussian probability density funtions:

```python
import numpy as np
from matplotlib import pyplot as plt
import math

def pdf(x, mu=0, sigma=1):
    "Calculates the normal distribution's probability density function (PDF)"
    
    term1 = 1.0/(math.sqrt(2*np.pi)*sigma)
    term2 = np.exp(-0.5*((x-mu)/sigma)**2)
    return term1*term2

x = np.arange(0, 100, 0.5)

#Returning the PDF
pdf1 = pdf(x, mu=5, sigma=2.5**0.5)
pdf2 = pdf(x, mu=10, sigma=6**0.5)

plt.plot(x, pdf1)
plt.plot(x, pdf2)

plt.title("Probability density functions")
plt.ylabel("p(x)")
plt.xlabel("Random variable x")
plt.legend(["pdf1 ~ N(5,2.5)","pdf2 ~ N(10,6)"], loc="upper right")
plt.ylim([0,0.5])
plt.xlim([0,20])

plt.show()
```

## Cumulative plots:

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.arange(1,11)
B = np.random.randn(10)
C = B.cumsum()

fig, (ax0,ax1) = plt.subplots(ncols=2, sharex=True, sharey=False, figsize=(10,5))

#A: via plt.step()

ax0.step(A, C, label="Cumulative sum")
ax0.scatter(A, B, label="Actual values")
ax0.set_ylabel("Y value")
ax0.legend(loc="upper right")
ax0.grid()

#B: via plt.plot()

ax1.plot(A, C, label="Cumulative sum")
ax1.scatter(A, B, label="Actual values")
ax1.set_ylabel("Y value")
ax1.legend(loc="upper right")
ax1.grid()

fig.text(0.5, 0.05, "Sample number", ha="center", va="center")
fig.text(0.5, 0.95, "Cumulative sum of 10 samples from a random normal distribution", ha="center", va="center")

plt.show()
```

## Absolute count:

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.arange(1,11)
B = np.random.randn(10)

plt.figure(figsize=(10,5))

plt.step(np.sort(B), A)
plt.ylabel("Sample count")
plt.xlabel("X value")
plt.title("Number of samples at a certain threshold")

plt.grid()
plt.show()
```

## Marker styles:

```python
import numpy as np
import matplotlib.pyplot as plt

markers = [
'.', # point
',', # pixel
'o', # circle
'v', # triangle down
'^', # triangle up
'<', # triangle_left
'>', # triangle_right
'1', # tri_down
'2', # tri_up
'3', # tri_left
'4', # tri_right
'8', # octagon
's', # square
'p', # pentagon
'*', # star
'h', # hexagon1
'H', # hexagon2
'+', # plus
'x', # x
'D', # diamond
'd', # thin_diamond
'|', # vline
]

plt.figure(figsize=(13, 10))
samples = range(len(markers))


for i in samples:
    plt.plot([i-1, i, i+1], [i, i, i], label=markers[i], marker=markers[i], markersize=10) 
#To make a three width line at the same level for each marker style

# Annotation    
plt.title("Matplotlib Marker styles", fontsize=20)
plt.ylim([-1, len(markers)+1])
plt.legend(loc="upper left")


plt.show()
```

## Line styles:

```python
import numpy as np
import matplotlib.pyplot as plt

linestyles = ["-.", "--", "None", "-", ":"]

plt.figure(figsize=(8, 5))
samples = range(len(linestyles)) #X values


for i in samples:
    plt.plot([i-1, i, i+1], [i, i, i], label='"%s"' %linestyles[i], linestyle=linestyles[i], lw=4) 

# Annotation   
plt.title("Matplotlib Line styles", fontsize=20)
plt.ylim([-1, len(linestyles)+1])
plt.legend(loc="upper left")


plt.show()
```
