# Getting data from CSV file:

```python
#set matplotlib to draw plots inline
%matplotlib inline

#Importing required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Passing list data to plot

x = [0,1,2,3,4,5]
y = [0,10,20,30,40,50]

plt.plot(x,y)
plt.show()
```

```python
x = [0,1,2,3,4,5]
y = [0,10,20,30,40,50]

plt.plot(x,y,marker="d")
plt.show()
```

```python
#Using numpy to process data and then pass it to matplotlib for plotting

x = np.arange(0,100)
y = x**2+7

plt.plot(x,y)
plt.show()
```

```python
#Plotting data from files read using pandas

data = pd.read_csv("gwp.csv")

data
```

```python
#Plotting pandas dataframe by passing columns and rows

plt.plot(data.Year, data.GWP, marker="D")
plt.show()
```

```python
#Data visualization by calling plot function directly on the dataframe

#Set year as index
data1 = data.set_index("Year")

data1
```

```python
#Creating a plot by calling plot function directly on the dataframe

data1.GWP.plot()
plt.show()
```

```python
data.describe()
```

```python
#Plotting a scatter plot

plt.figure(figsize=(10,6))

#Using scatter
plt.scatter(data.UnitPrice, data.Quantity)

plt.ylabel("Quantity")
plt.xlabel("Unit Price")

plt.show()
```

```python
#Plotting a different scatter plot with other columns

plt.figure(figsize=(10,6))

#Using scatter
plt.scatter(data.CustomerID, data.Quantity)

plt.ylabel("Quantity")
plt.xlabel("Customer ID")

plt.show()
```

```python
data.isnull().sum()
```

```python
#Plotting a different scatter plot with other columns

plt.figure(figsize=(10,6))

#Using scatter
plt.scatter(data.CustomerID, data.Quantity, data.UnitPrice)

plt.ylabel("Quantity")
plt.xlabel("Customer ID")

plt.show()
```

## Third part:

```python
#A line plot on our sales data

plt.figure(figsize=(10,6))

plt.plot(data.InvoiceDate, data.Quantity)

plt.xlabel("Invoice Date")
plt.ylabel("Quantity")

plt.show()
```

```python
#Using truth table

data[data.StockCode == 22827]
```

```python
data1 = data.loc[data.StockCode == 22827].sort_values("InvoiceDate")

data1
```

```python
plt.figure(figsize=(10,6))

plt.plot(data1.InvoiceDate, data1.Quantity)

plt.xlabel("Invoice Date")
plt.ylabel("Quantity")

plt.show()
```

```python
data.head()
```

```python
sales_by_product = pd.DataFrame({"total_sales":data.groupby(["StockCode"])["Quantity"].sum()}).reset_index().set_index(["StockCode"])

sales_by_product.head()
```

```python
#Plotting the dataset we made

plt.figure(figsize=(12,5))

chart = sales_by_product.total_sales.plot(kind="bar")

chart
plt.show()
```

```python
#Sorting top 10

top10 = sales_by_product.sort_values("total_sales", ascending=False)[:10]

chart = top10.total_sales.plot(kind="bar")
plt.show()
```

```python
#Sorting bottom 10

bot10 = sales_by_product.sort_values("total_sales", ascending=True)[:10]

chart = bot10.total_sales.plot(kind="bar")
plt.show()
```
