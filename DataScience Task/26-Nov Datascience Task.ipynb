## Replacing null values:

```python
data[data.id.isnull()] = data[data.id.isnull()].fillna(-1)
```

```python
data[data.diet.isnull()] = data[data.diet.isnull()].fillna("no fat")
```

```python
data[data.pulse.isnull()] = data[data.pulse.isnull()].fillna(75)
```

```python
data[data.time.isnull()] = data[data.time.isnull()].fillna("10 min")
```

```python
data[data.kind.isnull()] = data[data.kind.isnull()].fillna("Squats")
```

```python
data
```

## Plotting graph between two columns:

```python
plt.plot(data["Unnamed: 0"], data.pulse, marker="D")
plt.show()
```

## Adding a column:

```python
data["BMR"] = data["pulse"] * 20
```
