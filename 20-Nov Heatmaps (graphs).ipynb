## Using NumPy's histogram2d:

```python
from matplotlib import pyplot as plt
import numpy as np

mean = [0,0]
cov = [[0,1],[1,0]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

hist, xedges, yedges = np.histogram2d(x,y)

X, Y = np.meshgrid(xedges, yedges)
plt.imshow(hist)
plt.grid()
plt.colorbar()
plt.show()
```

```python
#Changing the interpolation

plt.imshow(hist, interpolation="nearest")
plt.grid()
plt.colorbar()
plt.show()
```

## Using hist2d from matplotlib:

```python
mean = [0,0]
cov = [[0,1],[1,0]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hist2d(x, y, bins=10)
plt.colorbar()
plt.grid()
plt.show()
```

## Using pcolor from matplotlib:

```python
mean = [0,0]
cov = [[0,1],[1,0]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

hist, xedges, yedges = np.histogram2d(x,y)

plt.pcolor(hist)
plt.colorbar()
plt.grid()
plt.show()
```

## Using matshow from matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

columns = ["A","B","C","D","E"]
rows = ["1","2","3","4","5"]

data = np.random.random((5,5))

fig = plt.figure()

ax = fig.add_subplot(111)

cax = ax.matshow(data, interpolation="nearest")
fig.colorbar(cax)

ax.set_xticklabels(['']+columns)
ax.set_yticklabels(['']+rows)

plt.show()
```

```python
#WAP to draw a line using given axis values taken from a text file

path = "20-Nov-test.txt"
file = open(path,"r")
data = file.read()
data = data.split("\n")

x = [int(row.split()[0]) for row in data]
y = [int(row.split()[1]) for row in data]

plt.plot(x, y, marker="x")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Title")

#plt.grid()
plt.show()
```

```python
#WAP to plot two or more lines with legends, different widths and colors
labels = ["Data 1", "Data 2"]
x = [1,2,3,4,5]
y_1 = [10,20,30,40,50]
y_2 = [30,40,50,60,70]

plt.plot(x, y_1, linewidth = 10)
plt.plot(x, y_2, linewidth = 5)

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Title")
plt.legend(labels, loc="upper left")

plt.show()
```

```python
#WAP to plot two or more lines with legends, different widths and colors and markers
labels = ["Data 1", "Data 2"]
x = [1,2,3,4,5]
y_1 = [10,20,30,40,50]
y_2 = [30,40,50,60,70]

plt.plot(x, y_1, marker="x", linewidth=10, markersize=15)
plt.plot(x, y_2, marker="D", linewidth=5, markersize=10)

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Title")
plt.legend(labels, loc="upper left")

plt.show()
```

```python
#WAP to plot several lines with different format styles in command using array

x = np.arange(0,5,0.2)

plt.plot(x, x, "g--") #green dashes
plt.plot(x, x**2, "bs") #blue squares
plt.plot(x, x**3, "r^") #red triangles

plt.xlabel("X axis")
plt.ylabel("Y axis")
#plt.yscale("log")
plt.title("Title")

plt.show()
```
